```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset, DataLoader
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pywt

# ---------- æ•°æ® ----------
DATA_PATH = r'D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC1_aging_durability_data_denoised.csv'
TRAIN_RATIO = 0.7

# ---------- åºåˆ— ----------
SEQ_LEN = 128
PRED_LEN = 1
SLIDE_STEP = PRED_LEN

# ---------- å°æ³¢ ----------
USE_WAVELET = False
WAVELET_NAME = 'db3'
WAVELET_LEVEL = 4
WAVELET_ALPHA = 0.5

# ---------- æ¨¡å‹ ----------
INPUT_SIZE = 8
HIDDEN_SIZE = 128
NUM_LAYERS = 2
DROPOUT = 0.0

# ---------- è®­ç»ƒ ----------
BATCH_SIZE = 64
EPOCHS = 100
LR = 0.001

# ---------- ç”»å›¾ ----------
MAX_PLOT_POINTS = 500
SAVE_FIG = False
SAVE_DIR = r"C:/Users/xiaoxiaoshadiao/Desktop/æ¯•ä¸šè®¾è®¡ç”»å›¾/"
save_pred_path = r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC1_pred_result.csv"
# ======================= STYLE =======================
sns.set_theme(style="whitegrid")
sns.set_context("talk")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


# =================== Wavelet =========================
def wavelet_denoise(signal):
    coeffs = pywt.wavedec(signal, WAVELET_NAME, level=WAVELET_LEVEL)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    threshold = WAVELET_ALPHA * sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs = [coeffs[0]] + [
        pywt.threshold(c, threshold, mode='soft') for c in coeffs[1:]
    ]
    return pywt.waverec(coeffs, WAVELET_NAME)[:len(signal)]


# =================== Dataset =========================
class TongjiDataset(Dataset):
    def __init__(self):
        df = pd.read_csv(DATA_PATH)

        df = df[['V', 'TinH2', 'TinAIR', 'I', 'J', 'DoutAIR', 'ToutH2', 'Dwat']]

        df = df.apply(pd.to_numeric, errors='coerce').dropna().reset_index(drop=True)

        # âœ… ä¿å­˜å®Œæ•´åŸå§‹ç”µå‹ï¼ˆç”¨äºç”»å›¾ï¼‰
        self.full_voltage = df['V'].values

        split = int(len(df) * TRAIN_RATIO)
        train_df, test_df = df.iloc[:split], df.iloc[split:]

        # âœ… ä¿å­˜è®­ç»ƒ / æµ‹è¯•çœŸå®ç”µå‹
        self.train_voltage = train_df['V'].values
        self.test_voltage = test_df['V'].values

        if USE_WAVELET:
            train_df = train_df.apply(wavelet_denoise)
            test_df = test_df.apply(wavelet_denoise)

        self.scaler = StandardScaler()
        train_df = self.scaler.fit_transform(train_df)
        test_df = self.scaler.transform(test_df)

        self.train_X, self.train_y = self.make_seq(train_df)
        self.test_X, self.test_y = self.make_seq(test_df)


    def make_seq(self, data):
        X, y = [], []
        for i in range(0, len(data) - SEQ_LEN - PRED_LEN, SLIDE_STEP):
            X.append(data[i:i + SEQ_LEN])
            y.append(data[i + SEQ_LEN:i + SEQ_LEN + PRED_LEN, 0])

        X = np.array(X, dtype=np.float32)
        y = np.array(y, dtype=np.float32)

        return torch.from_numpy(X), torch.from_numpy(y)


    def inverse(self, x):
        dummy = np.zeros((len(x), INPUT_SIZE))
        dummy[:, 0] = x
        return self.scaler.inverse_transform(dummy)[:, 0]


# =================== Model ===========================
class LSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(INPUT_SIZE, HIDDEN_SIZE, NUM_LAYERS,
                            batch_first=True, dropout=DROPOUT)
        self.fc = nn.Linear(HIDDEN_SIZE, PRED_LEN)

    def forward(self, x):
        o, _ = self.lstm(x)
        return self.fc(o[:, -1])


# =================== Trainer =========================
class Trainer:
    def __init__(self, model, loader, inverse, device):
        self.model = model
        self.loader = loader
        self.inverse = inverse
        self.device = device
        self.opt = optim.Adam(model.parameters(), lr=LR)
        self.loss_fn = nn.MSELoss()

    def run_epoch(self, train=True):
        self.model.train() if train else self.model.eval()
        preds, tars = [], []

        for x, y in self.loader:
            x = x.to(self.device)
            y = y.to(self.device)

            if train:
                self.opt.zero_grad()

            pred = self.model(x)
            loss = self.loss_fn(pred, y)

            if train:
                loss.backward()
                self.opt.step()

            preds.extend(self.inverse(pred.detach().cpu().numpy().flatten()))
            tars.extend(self.inverse(y.detach().cpu().numpy().flatten()))

        return preds, tars


# =================== Main ============================
def main():
    ds = TongjiDataset()

    train_loader = DataLoader(list(zip(ds.train_X, ds.train_y)),
                              batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(list(zip(ds.test_X, ds.test_y)),
                             batch_size=BATCH_SIZE)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = LSTM().to(device)
    trainer_train = Trainer(model, train_loader, ds.inverse, device)
    trainer_test = Trainer(model, test_loader, ds.inverse, device)

    best_rmse = float('inf')
    best_preds = None
    best_tars = None

    for e in range(EPOCHS):
        start = time.time()

        trainer_train.run_epoch(train=True)
        preds, tars = trainer_test.run_epoch(train=False)

        preds = np.array(preds)
        tars = np.array(tars)

        mae = mean_absolute_error(tars, preds)
        rmse = np.sqrt(mean_squared_error(tars, preds))
        mape = np.mean(np.abs((tars - preds) / (tars + 1e-8))) * 100
        r2 = r2_score(tars, preds)

        # âœ… è®°å½• RMSE æœ€ä¼˜ç»“æœ
        if rmse < best_rmse:
            best_rmse = rmse
            best_preds = preds.copy()
            best_tars = tars.copy()
            best_epoch = e + 1

            # âœ… ä¿å­˜æœ€ä¼˜æ¨¡å‹å‚æ•°
            torch.save(model.state_dict(), r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC1_best_model.pth")
            print(f"âœ… å·²ä¿å­˜æœ€ä¼˜æ¨¡å‹ï¼ˆEpoch {best_epoch}, RMSE={best_rmse:.6f}ï¼‰")

        print(
            f"Epoch {e + 1:03d} | "
            f"MAE {mae:.6f} | "
            f"MAPE {mape:.3f}% | "
            f"RMSE {rmse:.6f} | "
            f"RÂ² {r2:.6f} | "
            f"Time {time.time() - start:.2f}s"
        )

    print(f"\nâœ… æœ€ä¼˜ç»“æœå‡ºç°åœ¨ Epoch {best_epoch}")
    print(f"âœ… Best RMSE = {best_rmse:.6f}")


    # ================= ä¿å­˜è®­ç»ƒç»“æœ =================


    result_df = pd.DataFrame({
        "V_true": best_tars,
        "V_pred": best_preds
    })

    result_df.to_csv(save_pred_path, index=False)

    print(f"\nâœ… è®­ç»ƒç»“æœå·²ä¿å­˜è‡³ï¼š{save_pred_path}")
    print(f"âœ… ä¿å­˜æ ·æœ¬æ•°ï¼š{len(result_df)}")
    print("âœ… å‰ 5 è¡Œé¢„è§ˆï¼š")
    print(result_df.head())




    # ================= Plot =================
    fig, ax = plt.subplots(figsize=(14, 5), dpi=500)

    # ---------- 1ï¸âƒ£ å®Œæ•´åŸå§‹ç”µå‹ ----------
    ax.plot(
        range(len(ds.full_voltage)),
        ds.full_voltage,
        color='lightgray',
        linewidth=2.0,
        label='å®Œæ•´åŸå§‹ç”µå‹'
    )

    # ---------- 2ï¸âƒ£ è®­ç»ƒé›†çœŸå®ç”µå‹ ----------
    train_end = len(ds.train_voltage)
    ax.plot(
        range(train_end),
        ds.train_voltage,
        color='royalblue',
        linewidth=2.0,
        label='è®­ç»ƒé›†çœŸå®ç”µå‹'
    )

    # ---------- 3ï¸âƒ£ æµ‹è¯•é›†çœŸå®ç”µå‹ ----------
    ax.plot(
        range(train_end, train_end + len(ds.test_voltage)),
        ds.test_voltage,
        color='black',
        linewidth=2.0,
        label='æµ‹è¯•é›†çœŸå®ç”µå‹'
    )

    # ---------- 4ï¸âƒ£ æµ‹è¯•é›†é¢„æµ‹ç”µå‹ï¼ˆæœ€ä¼˜ RMSEï¼‰ ----------
    pred_start = train_end + SEQ_LEN
    pred_index = range(pred_start, pred_start + len(best_preds))

    ax.plot(
        pred_index,
        best_preds,
        color='firebrick',
        linewidth=2.0,
        alpha=0.85,
        label='æµ‹è¯•é›†é¢„æµ‹ç”µå‹'
    )

    ax.set_xlabel('æ ·æœ¬ç´¢å¼•', fontsize=16)
    ax.set_ylabel('è¾“å‡ºç”µå‹ (V)', fontsize=16)
    ax.set_title('IEEE FC1 ç”µå‹é¢„æµ‹ç»“æœï¼ˆæœ€ä¼˜ RMSEï¼‰',
                 fontsize=16, fontweight='bold')

    ax.legend(frameon=False, fontsize=16)
    ax.grid(False)

    plt.tight_layout()

    if SAVE_FIG:
        plt.savefig(f"{SAVE_DIR}/RUL_FC1_few_shot_10æ­¥.png", dpi=500)

    plt.show()


if __name__ == '__main__':
    main()
```

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.preprocessing import StandardScaler

# ===================== å…¨å±€é£æ ¼ =====================
sns.set_theme(style="white")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# ===================== è·¯å¾„ =====================
RAW_DATA_PATH = r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC1_aging_durability_data_denoised.csv"
MODEL_PATH = r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC1_best_model.pth"
SAVE_DIR = r"C:\Users\xiaoxiaoshadiao\Desktop\æ¯•ä¸šè®¾è®¡ç”»å›¾"

# ===================== æ¨¡å‹å‚æ•°ï¼ˆå¿…é¡»å’Œè®­ç»ƒä¸€è‡´ï¼‰ =====================
INPUT_SIZE = 8
HIDDEN_SIZE = 128
NUM_LAYERS = 2
SEQ_LEN = 128
PRED_LEN = 1

# ===================== RUL å‚æ•° =====================
RUL_START_INDEX = 7000        # âœ… å°šæœªå¤±æ•ˆçš„é¢„æµ‹èµ·ç‚¹ï¼ˆæ”¹å¤§ä¸€ç‚¹ å¦åˆ™çœ‹ä¸åˆ°å¤±æ•ˆ ä¹Ÿä¸èƒ½å¤ªå¤§ å¦åˆ™ä¼šä¸€è‡´å¤±æ•ˆï¼‰
INIT_WINDOW = 200
FAILURE_RATIOS = [0.03, 0.035, 0.04]
CONSECUTIVE_POINTS = 5
DELTA_T_MIN = 4.812           # FC1 æ—¶é—´åˆ†è¾¨ç‡ï¼ˆmin / sampleï¼‰

# ===================== æ¨¡å‹å®šä¹‰ =====================
class LSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(INPUT_SIZE, HIDDEN_SIZE, NUM_LAYERS,
                            batch_first=True)
        self.fc = nn.Linear(HIDDEN_SIZE, PRED_LEN)

    def forward(self, x):
        o, _ = self.lstm(x)
        return self.fc(o[:, -1])

# ===================== å·¥å…·å‡½æ•° =====================
def compute_v_init(voltage, window=200):
    v_init = np.mean(voltage[:window])
    print(f"\nâœ… åˆå§‹ç”µå‹ V_init = {v_init:.6f} V")
    return v_init


def find_failure_index(voltage, threshold, consecutive=5):
    count = 0
    for i, v in enumerate(voltage):
        if v < threshold:
            count += 1
            if count == consecutive:
                return i
        else:
            count = 0
    return None


def one_step_predict_with_true_v(model, data_scaled, start_idx, seq_len, max_steps):
    """
    æ¯ä¸€æ­¥éƒ½ä½¿ç”¨çœŸå®Vï¼ˆä¸æŠŠpredå–‚å›å»ï¼‰ï¼Œåªåšå•æ­¥é¢„æµ‹ï¼š
    - è¾“å…¥çª—å£å®Œå…¨æ¥è‡ªçœŸå®åºåˆ—ï¼ˆå·²æ ‡å‡†åŒ–ï¼‰
    - è¾“å‡ºä¸ºä¸‹ä¸€æ—¶åˆ»Vçš„é¢„æµ‹ï¼ˆæ ‡å‡†åŒ–åŸŸï¼‰
    """
    preds = []
    for k in range(max_steps):
        cur_seq = data_scaled[start_idx + k - seq_len : start_idx + k].copy()
        x = torch.tensor(cur_seq, dtype=torch.float32).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_v_scaled = model(x).cpu().numpy()[0, 0]
        preds.append(pred_v_scaled)
    return np.array(preds)


# ===================== ä¸»æµç¨‹ =====================
def main():
    # ---------- 1ï¸âƒ£ è¯»å–åŸå§‹æ•°æ® ----------
    df = pd.read_csv(RAW_DATA_PATH)
    features = df[['V','TinH2','TinAIR','I','J','DoutAIR','ToutH2','Dwat']].values
    v_true_full = features[:, 0]

    print(f"âœ… åŸå§‹æ•°æ®é•¿åº¦ï¼š{len(v_true_full)}")

    # ---------- 2ï¸âƒ£ æ ‡å‡†åŒ– ----------
    TRAIN_RATIO = 0.7
    split = int(len(features) * TRAIN_RATIO)

    scaler = StandardScaler()
    scaler.fit(features[:split])          # âœ… åªç”¨è®­ç»ƒæ®µæ‹Ÿåˆ
    features_scaled = scaler.transform(features)


    # ---------- 3ï¸âƒ£ åŠ è½½æ¨¡å‹ ----------
    global device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = LSTM().to(device)
    model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
    model.eval()

    print("âœ… å·²æˆåŠŸåŠ è½½ FC1 æœ€ä¼˜æ¨¡å‹")

    # ---------- 4ï¸âƒ£ æ„é€ çœŸå® RUL åºåˆ— ----------
    v_true_rul = v_true_full[RUL_START_INDEX:]
    print(f"âœ… RUL è¯„ä¼°èµ·ç‚¹ t0 = {RUL_START_INDEX}")

    # ---------- 5ï¸âƒ£ æ»šåŠ¨é¢„æµ‹ ----------
    pred_scaled = one_step_predict_with_true_v(
        model,
        features_scaled,
        start_idx=RUL_START_INDEX,
        seq_len=SEQ_LEN,
        max_steps=len(v_true_rul)
    )


    # ç”¨çœŸå®ç‰¹å¾çš„æ ‡å‡†åŒ–å€¼åšåŸºåº•ï¼Œåªæ›¿æ¢Vé‚£ä¸€åˆ—ä¸ºé¢„æµ‹å€¼ï¼Œå†æ•´ä½“ inverse_transform
    base = features_scaled[RUL_START_INDEX:RUL_START_INDEX + len(pred_scaled)].copy()
    base[:, 0] = pred_scaled
    v_pred_rul = scaler.inverse_transform(base)[:, 0]
    print("\nğŸ“Œ å‰10ä¸ªç‚¹å¯¹æ¯”ï¼š")
    for i in range(10):
        print(i, "true=", float(v_true_rul[i]), "pred=", float(v_pred_rul[i]))


    # ======== å…ˆç”»æ•´ä½“æ»šåŠ¨é¢„æµ‹ç»“æœï¼ˆä¸ä¾èµ–å¤±æ•ˆï¼‰ ========
    fig, ax = plt.subplots(figsize=(14,5), dpi=600)

    ax.plot(v_true_rul, color='black', linewidth=2, label='çœŸå®ç”µå‹')
    ax.plot(v_pred_rul, color='firebrick', linewidth=2, alpha=0.85, label='æ»šåŠ¨é¢„æµ‹ç”µå‹')

    ax.set_title(
        f'FC1 æ»šåŠ¨é¢„æµ‹ç»“æœï¼ˆèµ·ç‚¹ t0 = {RUL_START_INDEX}ï¼‰',
        fontsize=16, fontweight='bold'
    )
    ax.set_xlabel('é‡‡æ ·ç‚¹ç´¢å¼•')
    ax.set_ylabel('è¾“å‡ºç”µå‹ (V)')
    ax.legend(frameon=False)
    ax.grid(False)

    plt.tight_layout()
    plt.savefig(os.path.join(SAVE_DIR, f"FC1_roll_prediction_t0_{RUL_START_INDEX}.png"), dpi=600)
    plt.show()


    # ---------- 6ï¸âƒ£ RUL è®¡ç®— ----------
    v_init = compute_v_init(v_true_full, INIT_WINDOW)
    rul_records = []

    for ratio in FAILURE_RATIOS:
        v_ft = (1 - ratio) * v_init

        idx_true = find_failure_index(v_true_rul, v_ft)
        idx_pred = find_failure_index(v_pred_rul, v_ft)

        if idx_true is None or idx_pred is None:
            continue

        rul_ae = abs(idx_pred - idx_true)
        rul_re = rul_ae / idx_true
        time_ae = rul_ae * DELTA_T_MIN

        print("\n" + "="*60)
        print(f"ğŸ”» é˜ˆå€¼ {ratio*100:.1f}%")
        print(f"T_true = {idx_true}, T_pred = {idx_pred}")
        print(f"RUL_AE = {rul_ae} ç‚¹ ({time_ae:.2f} min)")
        print(f"RUL_RE = {rul_re:.4%}")

        rul_records.append([
            f"{ratio*100:.1f}%",
            idx_true,
            idx_pred,
            rul_ae,
            time_ae,
            rul_re
        ])

        # ---------- 7ï¸âƒ£ ç”»å›¾ ----------
        fig, ax = plt.subplots(figsize=(14,5), dpi=600)

        ax.plot(v_true_rul, color='black', linewidth=2, label='çœŸå®ç”µå‹')
        ax.plot(v_pred_rul, color='firebrick', linewidth=2, alpha=0.85, label='é¢„æµ‹ç”µå‹')

        ax.axhline(v_ft, linestyle='--', color='gray', label='å¤±æ•ˆé˜ˆå€¼')
        ax.axvline(idx_true, linestyle=':', color='royalblue', label='çœŸå®å¤±æ•ˆç‚¹')
        ax.axvline(idx_pred, linestyle=':', color='darkorange', label='é¢„æµ‹å¤±æ•ˆç‚¹')

        ax.set_title(f'FC1 RUL æ»šåŠ¨é¢„æµ‹ç»“æœï¼ˆé˜ˆå€¼ {ratio*100:.1f}%ï¼‰', fontsize=16, fontweight='bold')
        ax.set_xlabel('é‡‡æ ·ç‚¹ç´¢å¼•')
        ax.set_ylabel('è¾“å‡ºç”µå‹ (V)')
        ax.legend(frameon=False)
        ax.grid(False)

        plt.tight_layout()
        plt.savefig(os.path.join(SAVE_DIR, f"FC1_RUL_roll_{int(ratio*1000)}.png"), dpi=600)
        plt.show()

    # ---------- 8ï¸âƒ£ ä¿å­˜ç»“æœ ----------
    rul_df = pd.DataFrame(
        rul_records,
        columns=["å¤±æ•ˆé˜ˆå€¼","T_true","T_pred","RUL_AE(ç‚¹)","RUL_AE(åˆ†é’Ÿ)","RUL_RE"]
    )

    print("\nâœ… RUL ç»“æœæ±‡æ€»ï¼š")
    print(rul_df)


if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset, DataLoader
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pywt

# ---------- æ•°æ® ----------
DATA_PATH = r'D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC2_aging_durability_data_denoised.csv'
TRAIN_RATIO = 0.7

# ---------- åºåˆ— ----------
SEQ_LEN = 128
PRED_LEN = 1
SLIDE_STEP = PRED_LEN

# ---------- å°æ³¢ ----------
USE_WAVELET = False
WAVELET_NAME = 'db3'
WAVELET_LEVEL = 4
WAVELET_ALPHA = 0.5

# ---------- æ¨¡å‹ ----------
INPUT_SIZE = 8
HIDDEN_SIZE = 128
NUM_LAYERS = 2
DROPOUT = 0.0

# ---------- è®­ç»ƒ ----------
BATCH_SIZE = 64
EPOCHS = 30
LR = 0.001

# ---------- ç”»å›¾ ----------
MAX_PLOT_POINTS = 500
SAVE_FIG = False
SAVE_DIR = r"C:/Users/xiaoxiaoshadiao/Desktop/æ¯•ä¸šè®¾è®¡ç”»å›¾/"
save_pred_path = r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC2_pred_result.csv"
# ======================= STYLE =======================
sns.set_theme(style="whitegrid")
sns.set_context("talk")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


# =================== Wavelet =========================
def wavelet_denoise(signal):
    coeffs = pywt.wavedec(signal, WAVELET_NAME, level=WAVELET_LEVEL)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    threshold = WAVELET_ALPHA * sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs = [coeffs[0]] + [
        pywt.threshold(c, threshold, mode='soft') for c in coeffs[1:]
    ]
    return pywt.waverec(coeffs, WAVELET_NAME)[:len(signal)]


# =================== Dataset =========================
class TongjiDataset(Dataset):
    def __init__(self):
        df = pd.read_csv(DATA_PATH)

        df = df[['V', 'TinH2', 'TinAIR', 'I', 'J', 'DoutAIR', 'ToutH2', 'Dwat']]

        df = df.apply(pd.to_numeric, errors='coerce').dropna().reset_index(drop=True)

        # âœ… ä¿å­˜å®Œæ•´åŸå§‹ç”µå‹ï¼ˆç”¨äºç”»å›¾ï¼‰
        self.full_voltage = df['V'].values

        split = int(len(df) * TRAIN_RATIO)
        train_df, test_df = df.iloc[:split], df.iloc[split:]

        # âœ… ä¿å­˜è®­ç»ƒ / æµ‹è¯•çœŸå®ç”µå‹
        self.train_voltage = train_df['V'].values
        self.test_voltage = test_df['V'].values

        if USE_WAVELET:
            train_df = train_df.apply(wavelet_denoise)
            test_df = test_df.apply(wavelet_denoise)

        self.scaler = StandardScaler()
        train_df = self.scaler.fit_transform(train_df)
        test_df = self.scaler.transform(test_df)

        self.train_X, self.train_y = self.make_seq(train_df)
        self.test_X, self.test_y = self.make_seq(test_df)


    def make_seq(self, data):
        X, y = [], []
        for i in range(0, len(data) - SEQ_LEN - PRED_LEN, SLIDE_STEP):
            X.append(data[i:i + SEQ_LEN])
            y.append(data[i + SEQ_LEN:i + SEQ_LEN + PRED_LEN, 0])

        X = np.array(X, dtype=np.float32)
        y = np.array(y, dtype=np.float32)

        return torch.from_numpy(X), torch.from_numpy(y)


    def inverse(self, x):
        dummy = np.zeros((len(x), INPUT_SIZE))
        dummy[:, 0] = x
        return self.scaler.inverse_transform(dummy)[:, 0]


# =================== Model ===========================
class LSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(INPUT_SIZE, HIDDEN_SIZE, NUM_LAYERS,
                            batch_first=True, dropout=DROPOUT)
        self.fc = nn.Linear(HIDDEN_SIZE, PRED_LEN)

    def forward(self, x):
        o, _ = self.lstm(x)
        return self.fc(o[:, -1])


# =================== Trainer =========================
class Trainer:
    def __init__(self, model, loader, inverse, device):
        self.model = model
        self.loader = loader
        self.inverse = inverse
        self.device = device
        self.opt = optim.Adam(model.parameters(), lr=LR)
        self.loss_fn = nn.MSELoss()

    def run_epoch(self, train=True):
        self.model.train() if train else self.model.eval()
        preds, tars = [], []

        for x, y in self.loader:
            x = x.to(self.device)
            y = y.to(self.device)

            if train:
                self.opt.zero_grad()

            pred = self.model(x)
            loss = self.loss_fn(pred, y)

            if train:
                loss.backward()
                self.opt.step()

            preds.extend(self.inverse(pred.detach().cpu().numpy().flatten()))
            tars.extend(self.inverse(y.detach().cpu().numpy().flatten()))

        return preds, tars


# =================== Main ============================
def main():
    ds = TongjiDataset()

    train_loader = DataLoader(list(zip(ds.train_X, ds.train_y)),
                              batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(list(zip(ds.test_X, ds.test_y)),
                             batch_size=BATCH_SIZE)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = LSTM().to(device)
    trainer_train = Trainer(model, train_loader, ds.inverse, device)
    trainer_test = Trainer(model, test_loader, ds.inverse, device)

    best_rmse = float('inf')
    best_preds = None
    best_tars = None

    for e in range(EPOCHS):
        start = time.time()

        trainer_train.run_epoch(train=True)
        preds, tars = trainer_test.run_epoch(train=False)

        preds = np.array(preds)
        tars = np.array(tars)

        mae = mean_absolute_error(tars, preds)
        rmse = np.sqrt(mean_squared_error(tars, preds))
        mape = np.mean(np.abs((tars - preds) / (tars + 1e-8))) * 100
        r2 = r2_score(tars, preds)

        # âœ… è®°å½• RMSE æœ€ä¼˜ç»“æœ
        if rmse < best_rmse:
            best_rmse = rmse
            best_preds = preds.copy()
            best_tars = tars.copy()
            best_epoch = e + 1

            # âœ… ä¿å­˜æœ€ä¼˜æ¨¡å‹å‚æ•°
            torch.save(model.state_dict(), r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC2_best_model.pth")
            print(f"âœ… å·²ä¿å­˜æœ€ä¼˜æ¨¡å‹ï¼ˆEpoch {best_epoch}, RMSE={best_rmse:.6f}ï¼‰")

        print(
            f"Epoch {e + 1:03d} | "
            f"MAE {mae:.6f} | "
            f"MAPE {mape:.3f}% | "
            f"RMSE {rmse:.6f} | "
            f"RÂ² {r2:.6f} | "
            f"Time {time.time() - start:.2f}s"
        )

    print(f"\nâœ… æœ€ä¼˜ç»“æœå‡ºç°åœ¨ Epoch {best_epoch}")
    print(f"âœ… Best RMSE = {best_rmse:.6f}")


    # ================= ä¿å­˜è®­ç»ƒç»“æœ =================


    result_df = pd.DataFrame({
        "V_true": best_tars,
        "V_pred": best_preds
    })

    result_df.to_csv(save_pred_path, index=False)

    print(f"\nâœ… è®­ç»ƒç»“æœå·²ä¿å­˜è‡³ï¼š{save_pred_path}")
    print(f"âœ… ä¿å­˜æ ·æœ¬æ•°ï¼š{len(result_df)}")
    print("âœ… å‰ 5 è¡Œé¢„è§ˆï¼š")
    print(result_df.head())




    # ================= Plot =================
    fig, ax = plt.subplots(figsize=(14, 5), dpi=500)

    # ---------- 1ï¸âƒ£ å®Œæ•´åŸå§‹ç”µå‹ ----------
    ax.plot(
        range(len(ds.full_voltage)),
        ds.full_voltage,
        color='lightgray',
        linewidth=2.0,
        label='å®Œæ•´åŸå§‹ç”µå‹'
    )

    # ---------- 2ï¸âƒ£ è®­ç»ƒé›†çœŸå®ç”µå‹ ----------
    train_end = len(ds.train_voltage)
    ax.plot(
        range(train_end),
        ds.train_voltage,
        color='royalblue',
        linewidth=2.0,
        label='è®­ç»ƒé›†çœŸå®ç”µå‹'
    )

    # ---------- 3ï¸âƒ£ æµ‹è¯•é›†çœŸå®ç”µå‹ ----------
    ax.plot(
        range(train_end, train_end + len(ds.test_voltage)),
        ds.test_voltage,
        color='black',
        linewidth=2.0,
        label='æµ‹è¯•é›†çœŸå®ç”µå‹'
    )

    # ---------- 4ï¸âƒ£ æµ‹è¯•é›†é¢„æµ‹ç”µå‹ï¼ˆæœ€ä¼˜ RMSEï¼‰ ----------
    pred_start = train_end + SEQ_LEN
    pred_index = range(pred_start, pred_start + len(best_preds))

    ax.plot(
        pred_index,
        best_preds,
        color='firebrick',
        linewidth=2.0,
        alpha=0.85,
        label='æµ‹è¯•é›†é¢„æµ‹ç”µå‹'
    )

    ax.set_xlabel('æ ·æœ¬ç´¢å¼•', fontsize=16)
    ax.set_ylabel('è¾“å‡ºç”µå‹ (V)', fontsize=16)
    ax.set_title('IEEE FC2 ç”µå‹é¢„æµ‹ç»“æœï¼ˆæœ€ä¼˜ RMSEï¼‰',
                 fontsize=16, fontweight='bold')

    ax.legend(frameon=False, fontsize=16)
    ax.grid(False)

    plt.tight_layout()

    if SAVE_FIG:
        plt.savefig(f"{SAVE_DIR}/RUL_FC2_few_shot_10æ­¥.png", dpi=500)

    plt.show()


if __name__ == '__main__':
    main()
```

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.preprocessing import StandardScaler

# ===================== å…¨å±€é£æ ¼ =====================
sns.set_theme(style="white")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# ===================== è·¯å¾„ =====================
RAW_DATA_PATH = r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC2_aging_durability_data_denoised.csv"
MODEL_PATH = r"D:\xiaoxiaoshadiao\ems\data\processed\ieee\Durability_test_dataset\FC2_best_model.pth"
SAVE_DIR = r"C:\Users\xiaoxiaoshadiao\Desktop\æ¯•ä¸šè®¾è®¡ç”»å›¾"

# ===================== æ¨¡å‹å‚æ•°ï¼ˆå¿…é¡»å’Œè®­ç»ƒä¸€è‡´ï¼‰ =====================
INPUT_SIZE = 8
HIDDEN_SIZE = 128
NUM_LAYERS = 2
SEQ_LEN = 128
PRED_LEN = 1

# ===================== RUL å‚æ•° =====================
RUL_START_INDEX = 5000        # âœ… å°šæœªå¤±æ•ˆçš„é¢„æµ‹èµ·ç‚¹ï¼ˆæ”¹å¤§ä¸€ç‚¹ å¦åˆ™çœ‹ä¸åˆ°å¤±æ•ˆ  ä¹Ÿä¸å«©å¤ªå¤§  å¦åˆ™å…¨å¤±æ•ˆäº†ï¼‰
INIT_WINDOW = 200
FAILURE_RATIOS = [0.035, 0.04, 0.045]
CONSECUTIVE_POINTS = 5
DELTA_T_MIN = 4.805           # FC2 æ—¶é—´åˆ†è¾¨ç‡ï¼ˆmin / sampleï¼‰

# ===================== æ¨¡å‹å®šä¹‰ =====================
class LSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(INPUT_SIZE, HIDDEN_SIZE, NUM_LAYERS,
                            batch_first=True)
        self.fc = nn.Linear(HIDDEN_SIZE, PRED_LEN)

    def forward(self, x):
        o, _ = self.lstm(x)
        return self.fc(o[:, -1])

# ===================== å·¥å…·å‡½æ•° =====================
def compute_v_init(voltage, window=200):
    v_init = np.mean(voltage[:window])
    print(f"\nâœ… åˆå§‹ç”µå‹ V_init = {v_init:.6f} V")
    return v_init


def find_failure_index(voltage, threshold, consecutive=5):
    count = 0
    for i, v in enumerate(voltage):
        if v < threshold:
            count += 1
            if count == consecutive:
                return i
        else:
            count = 0
    return None


def one_step_predict_with_true_v(model, data_scaled, start_idx, seq_len, max_steps):
    """
    æ¯ä¸€æ­¥éƒ½ä½¿ç”¨çœŸå®Vï¼ˆä¸æŠŠpredå–‚å›å»ï¼‰ï¼Œåªåšå•æ­¥é¢„æµ‹ï¼š
    - è¾“å…¥çª—å£å®Œå…¨æ¥è‡ªçœŸå®åºåˆ—ï¼ˆå·²æ ‡å‡†åŒ–ï¼‰
    - è¾“å‡ºä¸ºä¸‹ä¸€æ—¶åˆ»Vçš„é¢„æµ‹ï¼ˆæ ‡å‡†åŒ–åŸŸï¼‰
    """
    preds = []
    for k in range(max_steps):
        cur_seq = data_scaled[start_idx + k - seq_len : start_idx + k].copy()
        x = torch.tensor(cur_seq, dtype=torch.float32).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_v_scaled = model(x).cpu().numpy()[0, 0]
        preds.append(pred_v_scaled)
    return np.array(preds)


# ===================== ä¸»æµç¨‹ =====================
def main():
    # ---------- 1ï¸âƒ£ è¯»å–åŸå§‹æ•°æ® ----------
    df = pd.read_csv(RAW_DATA_PATH)
    features = df[['V','TinH2','TinAIR','I','J','DoutAIR','ToutH2','Dwat']].values
    v_true_full = features[:, 0]

    print(f"âœ… åŸå§‹æ•°æ®é•¿åº¦ï¼š{len(v_true_full)}")

    # ---------- 2ï¸âƒ£ æ ‡å‡†åŒ– ----------
    TRAIN_RATIO = 0.7
    split = int(len(features) * TRAIN_RATIO)

    scaler = StandardScaler()
    scaler.fit(features[:split])          # âœ… åªç”¨è®­ç»ƒæ®µæ‹Ÿåˆ
    features_scaled = scaler.transform(features)


    # ---------- 3ï¸âƒ£ åŠ è½½æ¨¡å‹ ----------
    global device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = LSTM().to(device)
    model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
    model.eval()

    print("âœ… å·²æˆåŠŸåŠ è½½ FC2 æœ€ä¼˜æ¨¡å‹")

    # ---------- 4ï¸âƒ£ æ„é€ çœŸå® RUL åºåˆ— ----------
    v_true_rul = v_true_full[RUL_START_INDEX:]
    print(f"âœ… RUL è¯„ä¼°èµ·ç‚¹ t0 = {RUL_START_INDEX}")

    # ---------- 5ï¸âƒ£ æ»šåŠ¨é¢„æµ‹ ----------
    pred_scaled = one_step_predict_with_true_v(
        model,
        features_scaled,
        start_idx=RUL_START_INDEX,
        seq_len=SEQ_LEN,
        max_steps=len(v_true_rul)
    )


    # ç”¨çœŸå®ç‰¹å¾çš„æ ‡å‡†åŒ–å€¼åšåŸºåº•ï¼Œåªæ›¿æ¢Vé‚£ä¸€åˆ—ä¸ºé¢„æµ‹å€¼ï¼Œå†æ•´ä½“ inverse_transform
    base = features_scaled[RUL_START_INDEX:RUL_START_INDEX + len(pred_scaled)].copy()
    base[:, 0] = pred_scaled
    v_pred_rul = scaler.inverse_transform(base)[:, 0]
    print("\nğŸ“Œ å‰10ä¸ªç‚¹å¯¹æ¯”ï¼š")
    for i in range(10):
        print(i, "true=", float(v_true_rul[i]), "pred=", float(v_pred_rul[i]))


    # ======== å…ˆç”»æ•´ä½“æ»šåŠ¨é¢„æµ‹ç»“æœï¼ˆä¸ä¾èµ–å¤±æ•ˆï¼‰ ========
    fig, ax = plt.subplots(figsize=(14,5), dpi=600)

    ax.plot(v_true_rul, color='black', linewidth=2, label='çœŸå®ç”µå‹')
    ax.plot(v_pred_rul, color='firebrick', linewidth=2, alpha=0.85, label='æ»šåŠ¨é¢„æµ‹ç”µå‹')

    ax.set_title(
        f'FC2 æ»šåŠ¨é¢„æµ‹ç»“æœï¼ˆèµ·ç‚¹ t0 = {RUL_START_INDEX}ï¼‰',
        fontsize=16, fontweight='bold'
    )
    ax.set_xlabel('é‡‡æ ·ç‚¹ç´¢å¼•')
    ax.set_ylabel('è¾“å‡ºç”µå‹ (V)')
    ax.legend(frameon=False)
    ax.grid(False)

    plt.tight_layout()
    # plt.savefig(os.path.join(SAVE_DIR, f"FC2_roll_prediction_t0_{RUL_START_INDEX}.png"), dpi=600)
    plt.show()


    # ---------- 6ï¸âƒ£ RUL è®¡ç®— ----------
    v_init = compute_v_init(v_true_full, INIT_WINDOW)
    rul_records = []

    for ratio in FAILURE_RATIOS:
        v_ft = (1 - ratio) * v_init

        idx_true = find_failure_index(v_true_rul, v_ft)
        idx_pred = find_failure_index(v_pred_rul, v_ft)

        if idx_true is None or idx_pred is None:
            continue

        rul_ae = abs(idx_pred - idx_true)
        rul_re = rul_ae / idx_true
        time_ae = rul_ae * DELTA_T_MIN

        print("\n" + "="*60)
        print(f"ğŸ”» é˜ˆå€¼ {ratio*100:.1f}%")
        print(f"T_true = {idx_true}, T_pred = {idx_pred}")
        print(f"RUL_AE = {rul_ae} ç‚¹ ({time_ae:.2f} min)")
        print(f"RUL_RE = {rul_re:.4%}")

        rul_records.append([
            f"{ratio*100:.1f}%",
            idx_true,
            idx_pred,
            rul_ae,
            time_ae,
            rul_re
        ])

        # ---------- 7ï¸âƒ£ ç”»å›¾ ----------
        fig, ax = plt.subplots(figsize=(14,5), dpi=600)

        ax.plot(v_true_rul, color='black', linewidth=2, label='çœŸå®ç”µå‹')
        ax.plot(v_pred_rul, color='firebrick', linewidth=2, alpha=0.85, label='é¢„æµ‹ç”µå‹')

        ax.axhline(v_ft, linestyle='--', color='gray', label='å¤±æ•ˆé˜ˆå€¼')
        ax.axvline(idx_true, linestyle=':', color='royalblue', label='çœŸå®å¤±æ•ˆç‚¹')
        ax.axvline(idx_pred, linestyle=':', color='darkorange', label='é¢„æµ‹å¤±æ•ˆç‚¹')

        ax.set_title(f'FC2 RUL æ»šåŠ¨é¢„æµ‹ç»“æœï¼ˆé˜ˆå€¼ {ratio*100:.1f}%ï¼‰', fontsize=16, fontweight='bold')
        ax.set_xlabel('é‡‡æ ·ç‚¹ç´¢å¼•')
        ax.set_ylabel('è¾“å‡ºç”µå‹ (V)')
        ax.legend(frameon=False)
        ax.grid(False)

        plt.tight_layout()
        plt.savefig(os.path.join(SAVE_DIR, f"FC2_RUL_roll_{int(ratio*1000)}.png"), dpi=600)
        plt.show()

    # ---------- 8ï¸âƒ£ ä¿å­˜ç»“æœ ----------
    rul_df = pd.DataFrame(
        rul_records,
        columns=["å¤±æ•ˆé˜ˆå€¼","T_true","T_pred","RUL_AE(ç‚¹)","RUL_AE(åˆ†é’Ÿ)","RUL_RE"]
    )

    print("\nâœ… RUL ç»“æœæ±‡æ€»ï¼š")
    print(rul_df)


if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset, DataLoader
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pywt
import joblib
import os

# ---------- æ•°æ® ----------
DATA_PATH = r'D:\xiaoxiaoshadiao\ems\data\processed\lvdong\330_100_8_denoised.csv'
TRAIN_RATIO = 0.7

# ---------- åºåˆ— ----------
SEQ_LEN = 128
PRED_LEN = 1
SLIDE_STEP = PRED_LEN

# ---------- å°æ³¢ ----------
USE_WAVELET = False
WAVELET_NAME = 'db3'
WAVELET_LEVEL = 4
WAVELET_ALPHA = 0.5

# ---------- æ¨¡å‹ ----------
INPUT_SIZE = 8
HIDDEN_SIZE = 128
NUM_LAYERS = 2
DROPOUT = 0.0

# ---------- è®­ç»ƒ ----------
BATCH_SIZE = 64
EPOCHS = 100
LR = 0.001

# ---------- ä¿å­˜è·¯å¾„ ----------
SAVE_DIR = r"D:\xiaoxiaoshadiao\ems\data\processed\lvdong"
FIG_DIR = r"C:/Users/xiaoxiaoshadiao/Desktop/æ¯•ä¸šè®¾è®¡ç”»å›¾/"
os.makedirs(SAVE_DIR, exist_ok=True)

MODEL_SAVE_PATH = os.path.join(SAVE_DIR, "215_best_model.pth")
SCALER_SAVE_PATH = os.path.join(SAVE_DIR, "215_scaler.pkl")
METRIC_SAVE_PATH = os.path.join(SAVE_DIR, "215_metrics.csv")

# ======================= STYLE =======================
sns.set_theme(style="whitegrid")
sns.set_context("talk")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


# =================== Wavelet =========================
def wavelet_denoise(signal):
    coeffs = pywt.wavedec(signal, WAVELET_NAME, level=WAVELET_LEVEL)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    threshold = WAVELET_ALPHA * sigma * np.sqrt(2 * np.log(len(signal)))
    coeffs = [coeffs[0]] + [
        pywt.threshold(c, threshold, mode='soft') for c in coeffs[1:]
    ]
    return pywt.waverec(coeffs, WAVELET_NAME)[:len(signal)]


# =================== Dataset =========================
class TongjiDataset(Dataset):
    def __init__(self):
        df = pd.read_csv(DATA_PATH)
        df = df[['U', 'TinH2', 'ToutAIR', 'RH', 'PoutAIR', 'PinAIR', 'QAIR', 'TinAIR']]
        df = df.apply(pd.to_numeric, errors='coerce').dropna().reset_index(drop=True)

        self.full_voltage = df['U'].values

        split = int(len(df) * TRAIN_RATIO)
        train_df, test_df = df.iloc[:split], df.iloc[split:]

        self.train_voltage = train_df['U'].values
        self.test_voltage = test_df['U'].values

        if USE_WAVELET:
            train_df = train_df.apply(wavelet_denoise)
            test_df = test_df.apply(wavelet_denoise)

        self.scaler = StandardScaler()
        train_df = self.scaler.fit_transform(train_df)
        test_df = self.scaler.transform(test_df)

        self.train_X, self.train_y = self.make_seq(train_df)
        self.test_X, self.test_y = self.make_seq(test_df)

    def make_seq(self, data):
        X, y = [], []
        for i in range(0, len(data) - SEQ_LEN - PRED_LEN, SLIDE_STEP):
            X.append(data[i:i + SEQ_LEN])
            y.append(data[i + SEQ_LEN:i + SEQ_LEN + PRED_LEN, 0])
        return torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

    def inverse(self, x):
        dummy = np.zeros((len(x), INPUT_SIZE))
        dummy[:, 0] = x
        return self.scaler.inverse_transform(dummy)[:, 0]


# =================== Model ===========================
class LSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(INPUT_SIZE, HIDDEN_SIZE, NUM_LAYERS,
                            batch_first=True, dropout=DROPOUT)
        self.fc = nn.Linear(HIDDEN_SIZE, PRED_LEN)

    def forward(self, x):
        o, _ = self.lstm(x)
        return self.fc(o[:, -1])


# =================== Trainer =========================
class Trainer:
    def __init__(self, model, loader, inverse, device):
        self.model = model
        self.loader = loader
        self.inverse = inverse
        self.device = device
        self.opt = optim.Adam(model.parameters(), lr=LR)
        self.loss_fn = nn.MSELoss()

    def run_epoch(self, train=True):
        self.model.train() if train else self.model.eval()
        preds, tars = [], []

        for x, y in self.loader:
            x, y = x.to(self.device), y.to(self.device)

            if train:
                self.opt.zero_grad()

            pred = self.model(x)
            loss = self.loss_fn(pred, y)

            if train:
                loss.backward()
                self.opt.step()

            preds.extend(self.inverse(pred.detach().cpu().numpy().flatten()))
            tars.extend(self.inverse(y.detach().cpu().numpy().flatten()))

        return np.array(preds), np.array(tars)


# =================== Main ============================
def main():
    ds = TongjiDataset()

    train_loader = DataLoader(list(zip(ds.train_X, ds.train_y)),
                              batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(list(zip(ds.test_X, ds.test_y)),
                             batch_size=BATCH_SIZE)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = LSTM().to(device)

    trainer_train = Trainer(model, train_loader, ds.inverse, device)
    trainer_test = Trainer(model, test_loader, ds.inverse, device)

    best_rmse = float('inf')

    for e in range(EPOCHS):
        trainer_train.run_epoch(train=True)
        preds, tars = trainer_test.run_epoch(train=False)

        rmse = np.sqrt(mean_squared_error(tars, preds))
        mae = mean_absolute_error(tars, preds)
        mape = np.mean(np.abs((tars - preds) / (tars + 1e-8))) * 100
        r2 = r2_score(tars, preds)

        if rmse < best_rmse:
            best_rmse = rmse
            best_preds = preds.copy()
            best_tars = tars.copy()
            best_epoch = e + 1
            torch.save(model.state_dict(), MODEL_SAVE_PATH)

        print(f"Epoch {e+1:03d} | MAE {mae:.6f} | RMSE {rmse:.6f} | RÂ² {r2:.6f}")

    # âœ… ä¿å­˜ scaler
    joblib.dump(ds.scaler, SCALER_SAVE_PATH)

    # âœ… ä¿å­˜æŒ‡æ ‡
    metrics = pd.DataFrame([{
        "Best_Epoch": best_epoch,
        "MAE": mean_absolute_error(best_tars, best_preds),
        "RMSE": best_rmse,
        "MAPE(%)": np.mean(np.abs((best_tars - best_preds) / (best_tars + 1e-8))) * 100,
        "R2": r2_score(best_tars, best_preds)
    }])
    metrics.to_csv(METRIC_SAVE_PATH, index=False)

    print("\nâœ… æ¨¡å‹ã€Scalerã€æŒ‡æ ‡å·²å…¨éƒ¨ä¿å­˜")

    # ================= Plot =================
    fig, ax = plt.subplots(figsize=(14, 5), dpi=500)

    train_end = len(ds.train_voltage)
    pred_start = train_end + SEQ_LEN

    ax.plot(ds.full_voltage, color='lightgray', label='å®Œæ•´åŸå§‹ç”µå‹')
    ax.plot(range(train_end), ds.train_voltage, color='royalblue', label='è®­ç»ƒé›†çœŸå®ç”µå‹')
    ax.plot(range(train_end, train_end + len(ds.test_voltage)),
            ds.test_voltage, color='black', label='æµ‹è¯•é›†çœŸå®ç”µå‹')
    ax.plot(range(pred_start, pred_start + len(best_preds)),
            best_preds, color='firebrick', label='æµ‹è¯•é›†é¢„æµ‹ç”µå‹')

    ax.set_title('215é€šé“ç”µå‹é¢„æµ‹ç»“æœï¼ˆEAMFFï¼‰', fontsize=16, fontweight='bold')
    ax.set_xlabel('æ ·æœ¬ç´¢å¼•')
    ax.set_ylabel('è¾“å‡ºç”µå‹ (V)')
    ax.legend(frameon=False)
    ax.grid(False)

    plt.tight_layout()
    # plt.savefig(f"{FIG_DIR}/215é€šé“_EAMFF_é¢„æµ‹ç»“æœ.png", dpi=500)
    plt.show()


if __name__ == '__main__':
    main()
```
